```python
x = [1, 2, 3, 4]
y= [10, 20, 25, 30]

plt.plot(x,y)
plt.title("simple line chart")
plt.xlabel("time")
plt.ylabel("currency in the nearest million")
plt.grid(False)
plt.show()
```

```python
#pie chart
labels = ['python', 'java', 'c++', 'Ruby']
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.title("pie chart of programming languages")
plt.pie(sizes, labels=labels, colors = colors,  autopct= '%1.1f%%', startangle=0)
plt.axis('equal') #Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
import numpy as np

# Histogram
data = np.random.randn(10)
print(data)
plt.hist(data, bins=10, color= 'blue', alpha = 0.5) 
# plt.bar()
plt.title("Histogram of random data")

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Generate random.random data (numbers)
data1 = np.random.randn(100)
data2 = np.random.randn(100)
colors = np.random.rand(100)

#Create a scatter plot
plt.scatter(data1, data2, alpha=0.8, c=colors)
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [2, 3, 5, 7]

plt.plot(x,y1, label = 'Squared', marker='o')
plt.plot(x, y2, label = 'Prime', marker='x')
plt.title("Line chart with markers")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```
