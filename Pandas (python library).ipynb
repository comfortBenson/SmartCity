```python
if __name__== "__main__":
    file_path = 'Output.csv' #Replace with your csv file path
    data = pd.read_csv(file_path)
    print(data)
    df=pd.DataFrame(data)
    print(df)
```

```python
# check for null
print((df['name']).notnull())

def is__valid__email(email):
    import re
    return re.match(r"[^@]+@[^@]+\.[^@]+", email) is not None

df['validEmail'] = df['email'].apply(is__valid__email)
print((df['validEmail']))
df['RowValid'] = df['validEmail'] & df['name'].notnull() & df['age'].notnull()

print(df[['name', 'age', 'city', 'email', 'RowValid', 'validEmail']])
```

```python
df.loc[(df['name']).isna(), 'name'] = 'unknown'
print(df)
```
